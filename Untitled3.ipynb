```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# 1. Load the dataset
# Assuming your file is named 'pollution.csv'
df = pd.read_csv('/content/PRSA_data_2010.1.1-2014.12.31.csv')

# 2. Date-Time Parsing
# Combine year, month, day, and hour into a single 'date' column
df['date'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])

# Set this new 'date' column as the index (crucial for time-series)
df.set_index('date', inplace=True)

# Drop the original separate time columns and the 'No' column
df.drop(columns=['No', 'year', 'month', 'day', 'hour'], inplace=True)

# 3. Handling Missing Values (The NaNs in 'pm2.5')
# Strategy: Since the first 24 hours are usually NaN in this specific dataset,
# we often drop the initial rows or fill them.
# For the rest of the dataset, we use forward filling (propagate last valid observation forward).
df['pm2.5'].fillna(method='ffill', inplace=True)

# If there are still NaNs at the very start (where ffill can't look back), drop them
df.dropna(inplace=True)

# 4. Encoding Categorical Data
# The 'cbwd' column (Combined wind direction) is text (NE, NW, etc.).
# We encode it to integers.
encoder = LabelEncoder()
df['cbwd'] = encoder.fit_transform(df['cbwd'])

# 6. Final Inspection
print("Missing values remaining:", df.isnull().sum().sum())
print(df.head())
```
